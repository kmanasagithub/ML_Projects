```python
#Data Prepocessing - Data Cleaning
df.isnull().sum()
```

```python
#Replacing SalePrice empty values with their mean values
df["SalePrice"] = df["SalePrice"].fillna(df["SalePrice"].mean())
df
```

```python
# Drop records with null values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
# Drop Id as it doesn't contribute in price
df.drop("Id",axis=1,inplace=True)
df
```

```python
df["MSZoning"].unique()
```

```python
#Performing One - hot Encoding
cols = ["LotConfig","BldgType","MSZoning","Exterior1st"]
df = pd.get_dummies(df,columns=cols,drop_first=True,dtype="int64")
df
```

```python
## Train-Test Split
from sklearn.model_selection import train_test_split
X = df.drop(["SalePrice"],axis=1)
Y = df["SalePrice"]
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=0)
Y_train
```

```python
#feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
# Train Linear Regression Model
from sklearn.linear_model import LinearRegression
le = LinearRegression()
le.fit(X_train_scaled,y_train)
```

```python
# Model Evaluation
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
import numpy as np

y_pred = le.predict(X_test_scaled)

print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print(mean_absolute_percentage_error(y_test, y_pred))
```
